### Assignment 8:  Write a program to find the live weather report (temperature, wind speed, description, and weather) of a given city.

```python
import requests


def main():
	print("Enter the city name")
	city = input()
	#print(city)

	print('Displaying Weater report for: ' + city)

	url = 'https://wttr.in/{}'.format(city)
	res = requests.get(url)

	print(res.text)

if __name__ == '__main__':
	main()
```
